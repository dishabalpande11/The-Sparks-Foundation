# Task 2 Prediction using Unsupervised ML

 From the given ‘Iris’ dataset, predict the optimum number of clusters and epresent it visually.

### Loading Required Libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
import seaborn as sns

```

### Import data

```python
Data = pd.read_csv("Iris.csv")
```

```python
Data.head()
```

```python
Data.describe()
```

```python
Data.info()
```

```python
sns.pairplot(Data,hue='Species')
```

```python
sns.heatmap(Data.corr())
```

```python
X = Data.iloc[:,[0,1,2,3]].values
X
```

### K-Means Model

```python
kmeans = KMeans(n_clusters=3).fit(X)
```

```python
y_predict=kmeans.fit_predict(X)
y_predict
```

```python
plt.scatter(X[y_predict == 0,0],X[y_predict == 0,1],s=50,color='red', label='Iris-setosa')
plt.scatter(X[y_predict == 1,0],X[y_predict == 1,1],s=50,color='green',label='Iris-versicolour')
plt.scatter(X[y_predict == 2,0],X[y_predict == 2,2],s=50,color='yellow',label='Iris-verginica')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1],marker ='*', s = 80, c = 'black', label = 'Centroids')
plt.legend()
```

```python
plt.scatter(X[y_predict == 0,2],X[y_predict == 0,3],s=50,color='red',label='Iris-setosa')
plt.scatter(X[y_predict == 1,2],X[y_predict == 1,3],s=50,color='green',label='Iris-versicolour')
plt.scatter(X[y_predict == 2,2],X[y_predict == 2,3],s=50,color= 'yellow',label='Iris-verginica')
plt.scatter(kmeans.cluster_centers_[:,2],kmeans.cluster_centers_[:,3],marker="*",s=80,c='black',label='Centroids')
plt.legend()
```
